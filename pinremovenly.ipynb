```python
import numpy as np
import cv2

# Open the image.
img = cv2.imread('/content/output9.png')
H,W,_ = img.shape

# Load the mask.
mask = cv2.imread('/content/mask.png', 0)
mask = cv2.resize(mask,(W,H))

# Inpaint.
dst = cv2.inpaint(img, mask, 10, cv2.INPAINT_NS)

# Write the output.
cv2.imwrite('cat_inpainted.png', dst)
```
